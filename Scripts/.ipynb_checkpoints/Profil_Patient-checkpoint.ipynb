# Data Preprocessing

```r
# Aperçu des classes de colonnes
sapply(df_raw, class) %>% table()
```

```r
# Suppression colonne sans données
df_raw <- df_raw %>% select(-ALERTE_ORANGE_PREALABLE_A_1e_REHOSPITALISATION_POUR_IC, -ALERTE_ROUGE_PREALABLE_A_1e_REHOSPITALISATION_POUR_IC)
```

```r
# On Définit les types de nos colonnes
char_cols      <- c(
    "NUMERO_PATIENT"
)

integer_cols   <- c(
    "AGE",
    "DMS",
    "SCORE CHARLSON", 
    "NT-PRO_BNP_SORTIE",
    "FERRITINE",
    "DOSAGE_SORTIE_DIURETIC",
    "NBRE_REHOSPITALISATION_IC_12MOIS"
)

double_cols    <- c(
    "ADL",
    "IMC",
    "ALBUMINE"
)

percent_cols   <- c(
    "HbA1c", 
    "CST"
) 

bool_cols      <- c(
    "SEXE",
    "MODE_DE_VIE",
    "OBESITE",
    "SYNDROME_DEPRESSIF",
    "TROUBLES_COGNITIFS_SEVERES",
    "ATCD_CANCER",
    "CANCER_ACTIF",
    "ATCD_BPCO",
    "ATCD_HTA",
    "ATCD_DYSLIPIDEMIE",
    "ATCD_DIABETE",
    "ATCD_AVC",
    "ATCD_AOMI",
    "ATCD_FA/FLUTTER",
    "ATCD_AMYLOSE",
    "PM/DAI",
    "RAC_AU_MOINS_MODEREE",
    "IM_AU_MOINS_MODEREE",
    "IT_AU_MOINS_MODEREE",
    "CARENCE_MARTIALE_VRAIE",
    "CARENCE_MARTIALE_FONCTIONNELLE",
    "ANEMIE_HB<10g/l",
    "CARENCE_VITAMINE_B9",
    "CARENCE_VITAMINE_B12",
    "CARENCE_VITAMINE_D",
    "PRESENCE_IECi/ARA2",
    "PRESENCE_ACEi",
    "PRESENCE_BB-",
    "PRESENCE_MRA",
    "PRESENCE_SORTIE_ISGLT2",
    ">/=3_FANTASTIQUES_FEVG<50%",
    "PRESENCE_ISGLT2_FEVG>50%",
    "PRESENCE_SORTIE_DIURETIC",
    "PRESENCE_SORTIE_ANTICOAGULANT",
    "PRESENCE_SORTIE_AAP",
    "OUTCOME_REHOSPITALISATION_12MOIS"
)

ordinal_mappings <- list(
    DENUTRITION = c(
        "0" = "Risque dénutrition",
        "1" = "Dénutrition modérée",
        "2" = "Dénutrition sévère"
    ),
    ATCD_CARDIOPATHIE_ISCHEMIQUE = c(
        "0" = "NON", 
        "1" = "Traitement médical",
        "2" = "Angioplastie stent",
        "3" = "Pontage"
    ),
    FEVG_CLASSE = c(
        "0" = "Altérée (</=40%)",
        "1" = "Moyennement altérée (41-49%)",
        "2" = "Préservée (>/=50%)"
    ),
    STADE_IRC_DFG_SORTIE = c(
        "1" = "DFG>90 ml/min",
        "2" = "DFG=60-89 ml/min",
        "3A" = "45-59 ml/min",
        "3B" = "30-44 ml/min",
        "4" = "15-29 ml/min",
        "5" = "<14 ml/min"

    ),
    TYPE_DIURETIC = c(
        "0" = "FUROSEMIDE",
        "1" = "BUMETAMIDE"
    ),
    TYPE_SUIVI_SATELIA = c(
        "0" = "Patient", 
        "1" = "Aidant", 
        "2" = "IDEL",
        "3" = "Satelia"
    )
)

rehosp_cols    <- c(
  "REHOSPITALISATION_IC_1MOIS",
  "REHOSPITALISATION_IC_3MOIS",
  "REHOSPITALISATION_IC_6MOIS",
  "REHOSPITALISATION_IC_12MOIS",
  "REHOSPITALISATION_IC_>12MOIS"
)
```

```r
# Pipeline de typage
df_clean <- df_raw %>%
  # Caractères
  mutate(across(all_of(char_cols), as.character)) %>%
  # Doubles
  mutate(across(all_of(double_cols),
                ~ as.numeric( na_if(trimws(.x), "NC") ))) %>%
  # Entiers
  mutate(across(all_of(integer_cols),
                ~ as.integer(parse_number(trimws(.x), na = c("", "NC", "NA"))))) %>%
  # Pourcentages
  mutate(across(all_of(percent_cols),
                ~ parse_number(trimws(.x), locale=locale(decimal_mark=","), na=c("","NC"))/100)) %>%
  # Booléens → logique → factor Non/Oui
  mutate(across(all_of(bool_cols),
                ~ case_when(.x %in% c("1",1,TRUE) ~ TRUE,
                            .x %in% c("0",0,FALSE)~ FALSE, TRUE~NA))) %>%
  mutate(across(all_of(bool_cols),
                ~ factor(.x, levels=c(FALSE,TRUE), labels=c("Non","Oui")))) %>%
  # — Réhospitalisations → factor Non/Oui + NA en place
  mutate(across(all_of(rehosp_cols),
                ~ factor(
                    case_when(
                      .x == "1"              ~ "Oui",
                      .x == "0"              ~ "Non",
                      .x %in% c("NC","NA","") ~ NA_character_
                    ),
                    levels = c("Non","Oui")
                  )
         )) %>%
  # Ordinales via mapping
  { 
    df <- .
    for(col in names(ordinal_mappings)) {
      map <- ordinal_mappings[[col]]
      df <- df %>% mutate(
        !!col := {
          x   <- as.character(.data[[col]])
          lab <- map[x]                 # nommage par clé
          # lab est NA si x n’était pas un code valide (inclut NC/NA)
          factor(
            lab, 
            levels  = unname(map), 
            ordered = TRUE
          )
        }
      )
    }
    df
  }
```

```r
# 3. Sanity checks
glimpse(df_clean)
```

```r
tibble(
  Class = sapply(df_clean, function(x) class(x)[1])
) %>%
  count(Class, name = "Count")
```

```r
na_props <- function(cols) {
  props <- sapply(df_clean[cols], function(x) mean(is.na(x)))  
  props[props > 0]  
}

# 3. Appliquer à vos différents groupes de colonnes
integer_na  <- na_props(integer_cols)
double_na   <- na_props(double_cols)
percent_na  <- na_props(percent_cols)
bool_na     <- na_props(bool_cols)
rehosp_na   <- na_props(rehosp_cols)
ordinal_na  <- na_props(names(ordinal_mappings))

# 4. Afficher les résultats
print(list(
  integer   = integer_na,
  double    = double_na,
  percent   = percent_na,
  boolean   = bool_na,
  rehospitalisation = rehosp_na,
  ordinal   = ordinal_na
))
```

# Définition des Variables

```r
# Variables continues
cont_vars <- c(double_cols, integer_cols)

# Variables catégorielles
cat_vars  <- c(bool_cols, names(ordinal_mappings))

# Vairiable cible
rehosp_fac <- rehosp_cols

horizons   <- rehosp_cols[1:4]  
```

# All_Patient Univarié

```r
# Vérification pour modalité 1 mois
table(df_clean$REHOSPITALISATION_IC_1MOIS, useNA = "ifany")
```

```r
# Vérification pour modalité 3 mois
table(df_clean$REHOSPITALISATION_IC_3MOIS, useNA = "ifany")
```

```r
# Vérification pour modalité 12 mois
table(df_clean$REHOSPITALISATION_IC_12MOIS, useNA = "ifany")
```

```r
# Préparation du classeur Excel  
wb       <- createWorkbook()  
hdrStyle <- createStyle(textDecoration="Bold", halign="center", border="Bottom")

# Boucle sur chaque horizon
for (h in horizons) {
  # Créer la TableOne avec un max de détails
  tab <- CreateTableOne(
    vars        = c(cont_vars, cat_vars),
    strata      = h,
    data        = df_clean,
    factorVars  = cat_vars,
    includeNA   = TRUE,       # ajoute (Missing)
    addOverall  = TRUE,       # colonne All
    test        = TRUE,       # calcule p-values
    smd         = TRUE        # calcule standardized mean difference
  )

  # Afficher / récupérer sous forme de matrice
  mat <- print(
    tab,
    showAllLevels = TRUE,     # toutes les modalités
    quote         = FALSE,
    noSpaces      = FALSE,     # compact
    printToggle   = FALSE,    # renvoie la matrice
    smd           = TRUE,     # affiche la colonne SMD
    test.names    = TRUE,     # affiche le nom du test
    missing       = TRUE      # s’assure que (Missing) apparaît
  )

  # Transformer en data.frame et ajouter la colonne Variable
  df_tab <- as.data.frame(mat, stringsAsFactors = FALSE, check.names = FALSE) %>%
    rownames_to_column("Variable") %>%
    select(Variable, everything())

    df_tab$Variable <- gsub("^X(\\.\\d+)?$", "", df_tab$Variable)  # supprime les lignes comme "X.1"
    df_tab$Variable <- gsub("\\.{2,}", " ", df_tab$Variable)    # optionnel : nettoie les points
    df_tab$Variable <- trimws(df_tab$Variable)                  # optionnel : supprime les espaces

  # Écrire dans Excel
  sheet <- sub("REHOSPITALISATION_IC_", "", h) # p.ex. "1MOIS"
  addWorksheet(wb, sheetName = sheet)
    df_tab[is.na(df_tab)] <- "NA"
    writeDataTable(
    wb, sheet, df_tab,
    tableStyle = "TableStyleLight9",
    withFilter = TRUE,
    headerStyle = hdrStyle
  )
  freezePane(wb, sheet, firstRow = TRUE, firstCol = TRUE)
  setColWidths(wb, sheet, cols = 1:ncol(df_tab), widths = "auto")
}

#–– 4) Sauvegarde
output_file <- "../output/Table_univarie_complet.xlsx"
saveWorkbook(wb, file = output_file, overwrite = TRUE)
message("✅ Export complet dans ", normalizePath(output_file))
```
